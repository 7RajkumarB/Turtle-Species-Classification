```python
from google.colab import drive
drive.mount('/content/drive')

import os
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
from ultralytics import YOLO

# === SETUP DATA DIRECTORIES ===
train_dir = '/content/dataset/train'
val_dir = '/content/dataset/val'
test_dir = '/content/dataset/test'

!mkdir -p {train_dir} {val_dir} {test_dir}
!rsync -r '/content/drive/MyDrive/Turtle_classification_work/Sea_turtle_work/Sea_turtle_dataset(4 species)/Training/' {train_dir}/
!rsync -r '/content/drive/MyDrive/Turtle_classification_work/Sea_turtle_work/Sea_turtle_dataset(4 species)/Validation/' {val_dir}/
!rsync -r '/content/drive/MyDrive/Turtle_classification_work/Sea_turtle_work/Sea_turtle_dataset(4 species)/Testing/' {test_dir}/

print("Dataset structure:")
!apt install tree > /dev/null
!tree -L 2 /content/dataset

# Get class names
class_names = sorted(os.listdir(train_dir))
print("\nClass names:", class_names)

# === LOAD & TRAIN YOLOv8 CLASSIFIER ===
model = YOLO('yolov8n-cls.pt')  # Load pretrained classifier

# Set a custom project directory to avoid clutter in default location
model.args['project'] = '/content/yolo_results'
model.args['name'] = 'custom_classify'
model.args['exist_ok'] = True  # Don’t raise error if folder exists

# Train the model
results = model.train(
    data='/content/dataset',
    epochs=25,
    imgsz=224,
    batch=32,
    val=True,
    optimizer='AdamW',
    lr0=0.0001,
    patience=5,
    device='0',
)

# === EVALUATE ON TEST SET ===
test_images = []
y_true = []

for class_idx, class_name in enumerate(class_names):
    class_path = os.path.join(test_dir, class_name)
    if os.path.isdir(class_path):
        for img_name in os.listdir(class_path):
            if img_name.lower().endswith(('.jpg', '.jpeg', '.png')):
                test_images.append(os.path.join(class_path, img_name))
                y_true.append(class_idx)

print(f"Found {len(test_images)} test images.")

# Predict
y_pred = []
y_probs = []

start_time = time.time()
for img_path in test_images:
    result = model.predict(img_path, verbose=False)[0]
    y_pred.append(result.probs.top1)
    y_probs.append(result.probs.data.tolist())
end_time = time.time()

# Convert to arrays
y_true = np.array(y_true)
y_pred = np.array(y_pred)
y_probs = np.array(y_probs)

# === METRICS ===
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

acc = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred, average='weighted')
rec = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print(f"\n{' METRICS ':=^40}")
print(f"Test Accuracy: {acc * 100:.2f}%")
print(f"Precision: {prec * 100:.2f}%")
print(f"Recall: {rec * 100:.2f}%")
print(f"F1-Score: {f1 * 100:.2f}%")
print(f"Total inference time: {end_time - start_time:.2f} sec")
print(f"Average time/image: {(end_time - start_time)/len(test_images):.4f} sec")

# === CLASSIFICATION REPORT ===
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

# === CONFUSION MATRIX ===
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_true, y_pred), annot=True, fmt='d',
            xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# === ROC CURVE ===
plt.figure(figsize=(10, 8))
y_true_bin = label_binarize(y_true, classes=range(len(class_names)))

for i in range(len(class_names)):
    fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_probs[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{class_names[i]} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
